```python
# x-axis--->Categorical data
# y-axis--->Numerical data





import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.catplot(x='day',y='tip',data = tips)    
```

```python
sns.catplot(x='day',y='tip',kind='strip',data = tips)  #bydefault kind is strip
```

```python
sns.catplot(x='day',y='tip',kind='strip',jitter=1,data = tips)  #bydefault kind is strip
```

```python
sns.catplot(x='day',y='tip',kind='strip',jitter=0,data = tips)  #bydefault kind is strip
```

```python
sns.catplot(x='day',y='tip',kind='strip',jitter=2,data = tips)  
#Jitter is simply the addition of a small amount of horizontal (or vertical) 
# variability(space)to the data in order to ensure all data points are visible
```

```python
sns.catplot(x='day',y='tip',kind='swarm',data = tips)
#we can't use jitter function
```

```python
sns.catplot(x='day',y='tip',kind='swarm',data = tips,hue='sex')
```

```python
sns.swarmplot(x='day',y='tip',data = tips,hue='sex')
```

```python
sns.stripplot(x='day',y='tip',data = tips,hue='sex')
# we can't write stripplot and kind=strip at the same time
```
