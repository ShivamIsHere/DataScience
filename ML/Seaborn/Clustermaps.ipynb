```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
```

```python

flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
flights.pivot_table(index='year',columns='month',values='passengers',aggfunc='sum')
```

```python
x=flights.pivot_table(index='year',columns='month',values='passengers',aggfunc='sum')
sns.clustermap(x)
```

```python
sns.clustermap(x,metric='correlation')
```

```python
sns.clustermap(x,z_score=0,metric='correlation')
```

```python
sns.clustermap(x,z_score=0,metric='correlation',annot=True)
```

```python
sns.clustermap(x,z_score=0,metric='correlation',annot=True,row_cluster=False)
```

```python
sns.clustermap(x,z_score=0,metric='correlation',annot=True,row_cluster=False)
```
