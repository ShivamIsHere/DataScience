```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.catplot(x='smoker',y='total_bill',kind='bar',data=tips)
#black bar are confidience interval(mean)
```

```python
sns.catplot(x='smoker',y='total_bill',kind='bar',data=tips,hue='sex')
```

```python
sns.catplot(x='smoker',y='total_bill',kind='bar',estimator=np.median,data=tips,hue='sex')
```

```python
sns.catplot(x='smoker',y='total_bill',kind='bar',estimator=np.var,data=tips,hue='sex')
#var is variance
```

```python
sns.catplot(x='smoker',y='total_bill',kind='bar',estimator=np.std,data=tips,hue='sex')
```

```python
#count
```

```python
sns.catplot(x='sex',kind='count',data=tips)
```

```python
sns.catplot(x='sex',kind='count',data=tips,hue='smoker')
#catplot
```
