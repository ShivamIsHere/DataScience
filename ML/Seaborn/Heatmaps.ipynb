```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
flights.pivot_table(index='year',columns='month',values='passengers',aggfunc='sum')
```

```python
x=flights.pivot_table(index='year',columns='month',values='passengers',aggfunc='sum')
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x,cbar=False)
```

```python
sns.heatmap(x,cbar=False,linewidths=0.5)
```

```python
sns.heatmap(x,cbar=False,linewidths=0.5,annot=True)
# annot=annotation
```

```python
sns.heatmap(x,cbar=False,linewidths=0.5,annot=True,fmt='d')
# d is for integer
```

```python
sns.heatmap(x,cbar=False,linewidths=0.5,annot=True,fmt='d',cmap='viridis')
```

```python
sns.heatmap(x,cbar=False,linewidths=0.5,annot=True,fmt='d',cmap='summer')
```

```python
sns.heatmap(x,cbar=False,linewidths=0.5,annot=True,fmt='d',cmap='winter')
```
