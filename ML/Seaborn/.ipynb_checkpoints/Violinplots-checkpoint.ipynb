```python
#it give both the power of boxplot and kernel density estimation

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.violinplot(tips['total_bill'])
```

```python
sns.catplot(x='day',y='total_bill',kind='violin',data=tips)
```

```python
sns.catplot(x='day',y='total_bill',hue='sex',kind='violin',data=tips)
```

```python

sns.catplot(x='day',y='total_bill',hue='sex',kind='violin',data=tips,split=True)
#default value of split is false
```
