```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
```

```python
data = sns.load_dataset('tips')
```

```python
data.head()
```

```python
sns.relplot(x='total_bill',y='tip',data = data)
```

```python
sns.relplot(x='total_bill',y='tip',kind='scatter',data = data)
# kind can be either scatter or line is relplot
```

```python
 sns.relplot(x='total_bill',y='tip',hue='smoker',data = data)   #hue parameter
```

```python
 sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',data = data)   #style parameter
```

```python
 sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',size='size',data = data)   #size parameter
```

```python
 sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',size='size',sizes=(15,200),data = data)   #size parameter
```

```python
 sns.scatterplot(x='total_bill',y='tip',hue='smoker',style='sex',size='size',sizes=(15,200),data = data)  
    #rel and scatter are same thing
```
