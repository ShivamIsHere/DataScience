```python
# Displot combine 3things in one graph
# -->Histogram
# -->Kernel Density Enstimation(PDF)
# -->Rugplot




import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
titanic = sns.load_dataset('titanic')
```

```python
plt.style.use('fivethirtyeight')
```

```python
titanic.head()
```

```python
titanic['age'].fillna(titanic['age'].mean(),inplace=True)
```

```python
sns.displot(titanic['age'])
```

```python
sns.distplot(titanic['age'], kde=False)
```

```python
sns.distplot(titanic['age'], kde=False,bins=10)
```

```python
sns.distplot(titanic['age'], kde=False,bins=100)
```

```python
sns.distplot(titanic['age'], hist=False)
```

```python
sns.distplot(titanic['age'], kde=False,hist=False)
```

```python
sns.distplot(titanic['age'],hist=False, kde=False,rug=True)
```

```python
sns.distplot(titanic[titanic['survived']==1]['age'])
sns.distplot(titanic[titanic['survived']==0]['age'])
```
