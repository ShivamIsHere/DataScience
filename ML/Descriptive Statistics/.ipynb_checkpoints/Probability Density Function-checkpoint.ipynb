```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = sns.load_dataset('iris')
```

```python
data.head()
```

```python
data.rename(columns={'sepal_length':'SL','sepal_width':'SW','petal_length':'PL','petal_width':'PW'},inplace=True)
```

```python
iris_setosa=data[data['species']=='setosa']
iris_versicolor=data[data['species']=='versicolor']
iris_virginica=data[data['species']=='virginica']

#here we are creating 3different dataframe and storing values of previous data
```

```python
sns.distplot(iris_setosa['PL'])

#here curve is probability density function
#curve shows density of point at given range
```

```python
sns.FacetGrid(data, hue='species',height=6).map(sns.distplot,'PL').add_legend()
```

```python
sns.FacetGrid(data, hue='species',height=6).map(sns.distplot,'PL',kde=False).add_legend()
```

```python
sns.FacetGrid(data, hue='species',height=6).map(sns.distplot,'PW').add_legend()
```

```python
sns.FacetGrid(data, hue='species',height=6).map(sns.distplot,'PW',kde=False).add_legend()
```

```python
sns.FacetGrid(data, hue='species',height=6).map(sns.distplot,'SL').add_legend()
```

```python
sns.FacetGrid(data, hue='species',height=6).map(sns.distplot,'SW').add_legend()
```
