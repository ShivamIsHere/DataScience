```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days  = [1,2,3,4,5,6,7]
max_t = [50,59,55,57,54,51,56]
min_t = [43,42,40,44,41,47,49]
avg_t = [63,37,39,32,35,38,31]
```

```python
plt.xlabel("DAYS")
plt.ylabel("TEMPERATURE")
plt.title("WEATHER")
plt.plot(days,max_t, label="Max")    #label is used to represent which colour line represent which data
plt.plot(days,min_t, label="Min")
plt.plot(days,avg_t, label="Avg")
plt.legend()       #legend is used to name the line                  
```
