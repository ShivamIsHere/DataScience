```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals =[1400,600,300,410,250]
exp_labels =['Home Rent',"food","Phone/Internet Bill","Car","Other Utilities"]
```

```python
plt.pie(exp_vals,labels=exp_labels)   #we have to use keyword argument labels
```

```python
plt.pie(exp_vals,labels=exp_labels)  
plt.axis("equal")
```

```python
plt.pie(exp_vals,labels=exp_labels)  
```

```python
plt.pie(exp_vals,labels=exp_labels)  
plt.show()   #show will remove extra output text
```

```python
plt.pie(exp_vals,labels=exp_labels,radius=2)    #by default radius of size is 1
plt.show() 
```

```python
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct="%0.2f%%")    
plt.show() #0.2 will show two decimal point after percentage ..we can do 0.0f to hide decimal value
```

```python
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct="%0.2f%%",shadow=True)    
plt.show()    #it will show a small shadow around pie chart
```

```python
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct="%0.2f%%",shadow=True,explode=[0,0,0,0,0])    #nothing happens
plt.show()   #5zero's shows 5data
```

```python
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct="%0.2f%%",shadow=True,explode=[0,0.3,0.2,0,0.6])    #food pie comes out
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct="%0.2f%%",shadow=True,explode=[0,0.3,0.2,0,0.6],startangle=75)   
plt.show()    #home rent started at 75degree angle....using this we can rotate our pie chart
```
