```python
import pandas as pd
india_weather = pd.DataFrame({
    "city" : ["mumbai","delhi","banglore"],
    "temperature" : [32,45,30],
    "humidity" : [80,60,78]
})
india_weather
```

```python
import pandas as pd
us_weather = pd.DataFrame({
    "city" : ["new york","chicago","orlando"],
    "temperature" : [12,25,35],
    "humidity" : [30,40,75]
})
us_weather
```

```python
df=pd.concat([india_weather,us_weather]) 
df          #to add two data in single dataframe....
```

```python
df=pd.concat([india_weather,us_weather],ignore_index=True) 
df          #to ignore the continous indexing of the data frame
       # Go through pandas documantation on concatnation
```

```python
df=pd.concat([india_weather,us_weather],keys=["india","USA"]) 
df #it creates a addition index for subset of dataframe
#we have to remove ignore function inorder to use this addition indexing
```

```python
df.loc["india"]    #we can retrive a subset using loc
```

```python
temperature_weather = pd.DataFrame({
    "city" : ["mumbai","delhi","banglore"],
    "temperature" : [32,45,30]
})
temperature_weather
```

```python
windspeed_weather = pd.DataFrame({
    "city" : ["mumbai","delhi","banglore"],
    "humidity" : [80,60,78]
})
windspeed_weather
```

```python
df = pd.concat([temperature_weather,windspeed_weather],axis=1)
df   #axis=0 means it will append 2nd data as row 
#if we see axis=1 that means it will append 2nd data as column
```

```python
temperature_weather = pd.DataFrame({
    "city" : ["mumbai","delhi","banglore"],
    "temperature" : [32,45,30]
},index=[0,1,2])
temperature_weather
```

```python
windspeed_weather = pd.DataFrame({
    "city" : ["delhi","mumbai"],
    "humidity" : [80,60]
},index=[1,0])
windspeed_weather
```

```python
df = pd.concat([temperature_weather,windspeed_weather])
df   
```

```python
df = pd.concat([temperature_weather,windspeed_weather],axis=1)
df   #index is a way to aline rows from different dataframe
```
