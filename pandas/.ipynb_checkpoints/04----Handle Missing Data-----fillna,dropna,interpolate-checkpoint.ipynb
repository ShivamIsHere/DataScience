```python
df= pd.read_csv("weather.csv",parse_dates=["Date"]) #convert string in form of dates
type(df.Date[0])
```

```python
df.set_index('Date',inplace=True)
df #inplace is use to modify the original dataframe but if we won't use than instread it will written a new dataframe
```

```python
new_df=df.fillna(0)
new_df      #new is use to create new dataframe instead of modifing the current dataframe
```

```python
new_df =df.fillna({
    'MinTemp' : 0,
    'MaxTemp' : 0,
    'Rainfall' : 'No Rainfall',
    'Evaporation' : 0,
    'Sunshine' : 'cloudy'
})
new_df
```

```python
new_df = df.fillna(method='ffill')
new_df     #ffil means forward fill...which means just carry forward previous days value
```

```python
new_df = df.fillna(method='bfill') #checkout pandasfillna documentation on google 
new_df  #bfil means backward fill...which means just carry forward next days value
```

```python
new_df = df.fillna(method='bfill',axis="columns") #checkout pandasfillna documentation on google 
new_df #axis is used to fill in horizontal way
```

```python
new_df = df.fillna(method='ffill', limit=1) #checkout pandasfillna documentation on google 
new_df #limit=1 means copying valid value is one in the place of missing value
```

```python
new_df = df.fillna(method='ffill', limit=2) #checkout pandasfillna documentation on google 
new_df
```

```python
new_df= df.interpolate()
new_df    #Linear Interpolation is the technique of determining the values
# of the functions of any intermediate points when the values of two adjacent 
# points are known. Linear interpolation is basically the estimation of an 
# unknown value that falls within two known values. Linear Interpolation is 
# used in various disciplines like statistical, economics, price determination,
# etc. It is used to fill the gaps in the statistical data for the sake of 
# continuity of information
```

```python
new_df= df.interpolate(method="time")
new_df
```

```python
new_df=df.dropna() #it is used to drop all NaN value from the dataframe
new_df
```

```python
new_df=df.dropna(how="all") #it is used to drop only a row which have all the NaN value from the dataframe
new_df
```

```python
new_df=df.dropna(thresh=1) #it is used to drop a row which don't have atleast one valid datavalue
new_df
```

```python
new_df=df.dropna(thresh=2) #it is used to drop a row which don't have atleast two valid datavalue
new_df
```

```python
dt=pd.date_range("01-01-2001","01-09-2001")
idx=pd.DatetimeIndex(dt)
df=df.reindex(idx)
df    #it is used to show all the missing dates between two range
```
