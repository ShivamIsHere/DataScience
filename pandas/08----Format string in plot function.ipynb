```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,48]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,color='green',linewidth=5,linestyle='dashdot')
```

```python
plt.plot(x,y, 'g+-.')   #here g represent green colour line and -- double dash represent linestyle
```

```python
plt.plot(x,y, '--r*')    #order doesn't matter
#look out marker of matplotlib
```

```python
plt.plot(x,y, 'rD')   #sinle format string
```

```python
plt.plot(x,y, color='red',marker='D',linestyle='')    #three different argument
```

```python
plt.plot(x,y, color='#008040',marker='D',linestyle='')   #we can use hexadecimal for colour(00-red,80-green,40-blue)
```

```python
plt.plot(x,y, color='#008040',marker='D',linestyle='',markersize=15)   #to increase the marker size
```

```python
plt.plot(x,y, color='#008040',alpha=0.2) #alpha controls the transperency...basically scale of alpha is between---0to1
```
