```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company=['GOOGLE','AMAZONE','FACEBOOK','INSTAGRAM']
revenue=[90,136,89,27]
```

```python
plt.bar(company,revenue)   #if we do this function directly than string value will come in lower side
```

```python
xpos = np.arange(len(company))
xpos    #there are 4data so it printed 0to4
#we can do without using this functionxpos = np.arange(len(company))
xpos 
```

```python
plt.bar(xpos,revenue)   #it will show numeric number at the lower side 
```

```python
plt.xticks(xpos,company)    #it will show company at the bottom
plt.bar(xpos,revenue)
```

```python
plt.xticks(xpos,company)    #it will show company at the bottom
plt.ylabel("Revenue(bin)")
plt.title("US Tech Stock")
plt.bar(xpos,revenue)
```

```python
plt.xticks(xpos,company)    #it will show company at the bottom
plt.ylabel("Revenue(bin)")
plt.title("US Tech Stock")
plt.bar(xpos,revenue, label="Revenue")
plt.legend()
```

```python
company=['GOOGLE','AMAZONE','FACEBOOK','INSTAGRAM']
revenue=[90,136,89,27]
profit=[40,21,34,12]
```

```python
plt.xticks(xpos,company)    #it will show company at the bottom
plt.ylabel("Revenue(bin)")
plt.title("US Tech Stock")
plt.bar(xpos,revenue, label="Revenue")
plt.bar(xpos,profit,label="Profit")   #it will show data on same bar,this is because xbar position is same
plt.legend()
```

```python
plt.xticks(xpos,company)    #it will show company at the bottom
plt.ylabel("Revenue(bin)")
plt.title("US Tech Stock")
plt.bar(xpos-0.2,revenue, label="Revenue")
plt.bar(xpos+0.2,profit,label="Profit")   #-0.2 and +0.2 is used to create another bar chart side by side(seprated)


plt.legend()
```

```python
# xpos = np.arange(len(company))
# xpos   #we created numpy so that we can add and subtract 0.2 from 0,1,2,3
```

```python
plt.xticks(xpos,company)    
plt.ylabel("Revenue(bin)")
plt.title("US Tech Stock")
plt.bar(xpos-0.2,revenue,width=0.4, label="Revenue")
plt.bar(xpos+0.2,profit,width=0.4,label="Profit")   #width is used to seprate both bar chart of given length


plt.legend()
```

```python
plt.xticks(xpos,company)   
# plt.ylabel("Revenue(bin)")
plt.title("US Tech Stock")
plt.barh(xpos-0.2,revenue, label="Revenue")
plt.barh(xpos+0.2,profit,label="Profit")   #barh is used to represent data in horizontal bar chart
# we can't use width while using barh chart
#here company name will be of no use


plt.legend()
```

```python
plt.yticks(xpos,company)    #here we will use yticks show that we can represent company name in y-axis





# plt.ylabel("Revenue(bin)")
plt.title("US Tech Stock")
plt.barh(xpos-0.2,revenue, label="Revenue")
plt.barh(xpos+0.2,profit,label="Profit")   #barh is used to represent data in horizontal bar chart
# we can't use width while using barh chart
#here company name will be of no use


plt.legend()
```
