```python
import pandas as pd

def convert_phone_cell(cell):
    if cell=="01944-369967":
        return '8298605141'
    return cell
def convert_city_cell(cell):
    if cell=="St. Stephens":
        return 'patna'
    return cell
df=pd.read_excel('DataSheets.xlsx', "CompanyDetails" , converters ={
    'phone': convert_phone_cell,
    'city' : convert_city_cell
})
df
```

```python
df.to_excel("new.xlsx",sheet_name="CompanyDetails", startrow=1,startcol=2,index=False)
```

```python
df_stocks = pd.DataFrame({
    'tickers': ['GOOGL', 'WMT', 'MSFT'],
    'price': [845, 65, 64 ],
    'eps': [27.82, 4.61, 2.12],
    'pe': [30.37, 14.26, 30.97]
})

df_weather = pd.DataFrame({
    'day': ['1/1/2017','1/2/2017','1/3/2017'],
    'temperature': [32,35,28],
    'event': ['Rain', 'Sunny', 'Snow']
})
#writing two datasheet in one excel file
```

```python
with pd.ExcelWriter('stocks_weather.xlsx') as writer:
    df_stocks.to_excel(writer, sheet_name="stocks")
    df_weather.to_excel(writer, sheet_name="weather")
    
```
