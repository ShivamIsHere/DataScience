```python
import pandas as pd
import numpy as np
df=pd.read_csv('Weather_Data.csv')
df
```

```python
new_df = df.replace('-99999',np.NaN)
new_df
```

```python
new_df = df.replace(["-99999" , "-88888"],np.NaN)
new_df     #we can replace multiple value using dictionary
```

```python
new_df = df.replace({
    'temperature':"-99999",
    'windspeed':"-88888",
    'event':'0'
},np.NaN)
new_df    #inorder to avoid replacement of 0 in windspeed we use this kind of dictionary
```

```python
new_df = df.replace({
    "-99999": np.NaN,
    "-88888": np.NaN,
    'No Event':'Snow'
})
new_df   #we can use dictionary like this to replace none NaN value
```

```python
new_df = df.replace('[A-Za-z]','',regex=True)
new_df  #it will repace all non integer value of the dataframe 
        #it will skip the integer value where data is the Alpha-Numeric
#take a tutorial 
```

```python
new_df = df.replace({
    'temperature': '[A-Za-z]',
    'windspeed': '[A-Za-z]'
},'',regex=True)
new_df
```
